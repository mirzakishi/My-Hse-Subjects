### Центральная предельная теорема своими руками

#### Импортируем необходимые библиотеки

```python
import numpy as np
import seaborn as sns
import statistics as stat
```

#### Сгенерируем выборку объёма 1000 из  нормального распределения

```python
population = np.random.normal(loc=1, scale=2, size=10000) # среднее 1, стандартное отклонение 2
```

```python
sns.distplot(population)
```

#### Подыборка из нормально распределенной популяции

```python
x_sample = np.random.choice(population, size=10, replace=False)
x_sample 
```

```python
stat.mean(x_sample) # Среднее значение подвыборки
```

```python
# Функция подбирает n(количество) выборка из общей популяции, и каждый выборка состоит n случайных величин(размер)
def sample_mean_calculator(population_array, sample_size, n_samples):
  sample_means = []
  for i in range(n_samples):
    sample = np.random.choice(population_array, size=sample_size, replace=False)
    sample_mean = stat.mean(sample)
    sample_means.append(sample_mean)
  return sample_means
```

```python
sns.distplot(sample_mean_calculator(population, 10, 10), color='blue')
```

```python
sns.distplot(sample_mean_calculator(population, 100, 10), color='blue')
```

```python
sns.distplot(sample_mean_calculator(population, 1000, 100), color='blue')
```

#### Вывод

#### Из картинки видно что с увеличением количества и размер  выборки распределение становится более нормальном, что подтверждается "Центральная предельная теорема".
