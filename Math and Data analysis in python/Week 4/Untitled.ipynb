#### Example of distribution

```python
import numpy as np
from matplotlib import pyplot
from numpy.random import normal
from numpy import hstack
import statistics as stat
import scipy.stats as sts
```

```python
# generate a sample
norm_rv = sts.norm(0, 1)
sample = norm_rv.rvs(100)
```

```python
# plot the histogram
pyplot.hist(sample, bins=80, density=True, color='blue')
pyplot.show()
```

```python
x = np.linspace(-4,4,1000) # sample
pdf = norm_rv.pdf(x) # probabilities 

pyplot.hist(sample, bins=100, density=True, color='blue')
pyplot.plot(x, pdf, color='red',  alpha=0.8) 
```
